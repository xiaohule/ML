```python
pays_set={'France', 'Bénin', "Côte d'ivoire", 'France', 'Madagascar'}
def country_filter(p):
    if p not in pays_set:
        return np.nan
    return p
    
```

```python
data.pays=data.pays.apply(country_filter)
```

```python
data
```

```python
def convert_height(height):
    found = re.search('\d\.\d{2}', height)
    if found is None:
        found2=re.search('\d{3}', height)
        if found2 is None:
            print('{} n\'est pas au bon format. Il sera ignoré.'.format(height))
            return np.NaN
        else:
            print('{} n\'est pas au bon format. Il sera converti.'.format(height))
            value=height[:3]
            value=float(value)
            value=value/100
            return(value)
    else:
        value = height[:-1] # on enlève le dernier caractère, qui est 'm'
        return float(value)
```

```python
data['taille'] = [convert_height(t) for t in data['taille']]
```

```python
data['taille'] = [t if t<3 else np.NaN for t in data['taille']]
```

```python
data
```

```python
def fill_missing_height(height, replacement):
    if pd.isnull(height):
        print('Imputation par la moyenne : {}'.format(replacement))
        return replacement
    return height
```

```python
mean_height = data['taille'].mean()
data['taille'] = [fill_missing_height(t, mean_height) for t in data['taille']]
```

```python
data.date_naissance=pd.to_datetime(data.date_naissance,format="%d/%m/%Y",errors="coerce")
```

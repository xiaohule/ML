```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
#load dataset
data=pd.read_csv("hubble_data.csv")
#view dataset
data.head()
```

```python
#perform linear regression using sklearn
reg=LinearRegression().fit(data.distance.values.reshape(-1,1),data.recession_velocity.values)
#create common figure
fig=plt.figure()
#plot datapoints
plt.scatter(data.distance,data.recession_velocity)
#plot linear regression over datapoints range
x=np.linspace(0,2,10)
plt.plot(x,reg.predict(x.reshape(-1,1)))
```

```python
print("The recession velocity of a galaxy is linked to the distance with earth according to the relation:\nrecession_velocity=distance x",reg.coef_[0],reg.intercept_)
```

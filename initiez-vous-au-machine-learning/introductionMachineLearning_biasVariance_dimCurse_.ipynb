Le manque de donnes ncessaire  l'apprentissage du modle explose trs vite lorsqu'on augmente la dimension des donnes en entre, c'est pourquoi on surnomme ce phnomne curse of dimensionality.

Pour visualiser ce phnomne autrement, imaginez que vous devez traiter un problme qui possde N features et chacune peut possder 10 valeurs diffrentes. Notre algorithme d'apprentissage doit pouvoir distinguer parmi  1010...=10N configurations diffrentes. On voit rapidement que plus N grandi (la dimension des donnes), plus un algorithme aura besoin d'exemples pour pouvoir apprendre correctement le phnomne. C'est pour cette raison qu'on dit qu'augmenter la dimension des donnes d'entre est un flau pour le machine learning.

### The Inconvenient Truth About Data Science

- Data is never clean.

- You will spend most of your time cleaning and preparing data.

- 95% of tasks do not require deep learning.

- In 90% of cases generalized linear regression will do the trick.

- Big Data is just a tool.

- You should embrace the Bayesian approach.

- No one cares how you did it.

- Academia and business are two different worlds.

- Presentation is key - be a master of Power Point.

- All models are false, but some are useful.

- There is no fully automated Data Science. You need to get your hands dirty.

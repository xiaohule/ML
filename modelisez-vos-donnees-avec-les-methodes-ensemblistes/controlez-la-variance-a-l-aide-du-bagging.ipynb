```python
%matplotlib inline
from sklearn import ensemble, datasets, model_selection
from mglearn import plot_2d_separator, plot_interactive_tree, tools
```

```python
X, y = datasets.make_moons(noise=0.2)
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y, stratify=y)
```

```python
bagging = ensemble.BaggingClassifier(n_estimators=5, n_jobs=-1)
bagging.fit(X_train, y_train)

fig, axes = plt.subplots(2, 3, figsize=(20, 10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), bagging.estimators_)):
    ax.set_title("Tree {}".format(i))
    plot_interactive_tree.plot_tree_partition(X_train, y_train, tree, ax=ax)
    plot_2d_separator.plot_2d_separator(bagging, X_train, fill=True, ax=axes[-1, -1], alpha=.4)
    axes[-1, -1].set_title("Bagging")
    tools.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt,pi,exp

domaine=range(-100,100)
mu=0
sigma=20

f= lambda x: 1/sqrt(2*pi*sigma**2)*exp(-(x-mu)**2/(2*sigma**2))

y=[f(x) for x in domaine]
plot=plt.plot(domaine,y)
```
